```python
import pandas as pd
salespreddf = pd.read_csv('/content/drive/MyDrive/Coding Dojo/Data Sets/sales_predictions.csv')
salespreddf.head()
```

```python
salespred1df = salespreddf
salespred1df['Item_Weight']=salespred1df['Item_Weight'].fillna(salespred1df['Item_Weight'].mean())
```

```python
salespred2df = salespred1df.drop(columns=['Item_Weight','Outlet_Size','Outlet_Establishment_Year'])
salespred2df.head()
```

```python
salespred2df['Item_Fat_Content'].replace(to_replace='LF',value='Low Fat', inplace=True)
```

```python
salespred2df['Item_Fat_Content'].replace(to_replace='low fat',value='Low Fat', inplace=True)
```

```python
salespred2df['Item_Fat_Content'].replace(to_replace='reg', value='Regular', inplace=True)
```

```python
salespred2df.info()
```

```python
# Spend time transforming your categorical variables into numbers and use dummy encoding where appropriate.
#transform Item_Fat_Content
#transform Item_Type
spdf = pd.get_dummies(salespred2df, columns=['Item_Identifier','Item_Fat_Content','Item_Type',
                                             'Outlet_Identifier','Outlet_Location_Type','Outlet_Type'], drop_first=True)
spdf.head()
```

```python
# Assign the "Item_Outlet_Sales" column as your target and the rest of the variables as your features matrix. 
y = spdf['Item_Outlet_Sales']
X = spdf.drop(['Item_Outlet_Sales'], axis = 1)
```

```python
# Be sure to train test split your data set!
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Build a linear regression model.
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Evaluate your test set results using r2.
test_score = reg.score(X_test, y_test)
print(test_score)
```

```python
train_score = reg.score(X_train, y_train)
print(train_score)
```

```python
# Evaluate your test set results using RMSE
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
print('Training RMSE:', np.sqrt(mean_squared_error(y_train, reg.predict(X_train))))
print('Testing RMSE:', np.sqrt(mean_squared_error(y_test, reg.predict(X_test))))
```
